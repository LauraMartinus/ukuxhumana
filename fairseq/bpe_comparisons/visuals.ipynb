Creating graphs to view the results of the BPE study

```python
bpe_values = [4000,8000,12000,16000,20000,24000,28000,32000,36000,40000]
```

```python
af_results = [25.04,23.20,20.88,23.06,13.46,22.80,20.91,20.61,16.28,19.00]
nso_results = [12.18,8.73,10.14,10.07,8.68,6.98,8.21,7.27,8.65,7.50]
tn_results = [24.34,21.28,25.62,26.03,22.09,25.34,25.10,26.18,25.19,26.36]
ts_results = [33.21,34.94,35.75,36.67,37.45,35.46,35.07,35.44,37.47,35.44] 
zu_results = [1.79,1.66,1.27,0.85,1.07,1.33,0.96,0.40,0.57,0.35]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(bpe_values, af_results)
plt.title('BLEU score of Afrikaans on varying BPE values');
```

```python
plt.plot(bpe_values, nso_results)
plt.title('BLEU score of Northern Sotho on varying BPE values');
```

```python
plt.plot(bpe_values, tn_results)
plt.title('BLEU score of Setswana on varying BPE values');
```

```python
plt.plot(bpe_values, ts_results)
plt.title('BLEU score of Xitsonga on varying BPE values');
```

```python
plt.plot(bpe_values, zu_results)
plt.title('BLEU score of Zulu on varying BPE values');
```
